```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pickle

# Path to your dataset
dataset_path = './archive/asl_dataset/asl_dataset/'

# Initialize lists for images and labels
images = []
labels = []

# Load images and labels
for label in os.listdir(dataset_path):
    label_folder = os.path.join(dataset_path, label)
    if os.path.isdir(label_folder):
        for img_file in os.listdir(label_folder):
            img_path = os.path.join(label_folder, img_file)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  # Load as grayscale
            if img is None:  # Check if the image is loaded successfully
                print(f"Failed to load image: {img_path}")
                continue
            img = cv2.resize(img, (64, 64))  # Resize for consistency
            images.append(img)
            labels.append(label)

# Convert lists to NumPy arrays
images = np.array(images) / 255.0  # Normalize pixel values to [0, 1]
labels = np.array(labels)

# Encode labels (convert letters/numbers to integers)
label_encoder = LabelEncoder()
labels_encoded = label_encoder.fit_transform(labels)

# Check encoded labels
print(f"Encoded labels (first 5): {labels_encoded[:5]}")

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    images, labels_encoded, test_size=0.2, random_state=42
)

# Reshape for CNN input
X_train = X_train.reshape(-1, 64, 64, 1)
X_test = X_test.reshape(-1, 64, 64, 1)

# Save the label encoder to a file
with open('label_encoder.pkl', 'wb') as f:
    pickle.dump(label_encoder, f)

# Check the shape of the images array
print(f"Images shape: {images.shape}")

# Display a sample image
plt.imshow(X_train[0].reshape(64, 64), cmap='gray')
plt.title(f"Label: {label_encoder.inverse_transform([y_train[0]])[0]}")
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import numpy as np

# Define the CNN model
model = Sequential([
    # Input Layer
    Input(shape=(64, 64, 1)),  # For grayscale images (adjust if using RGB)

    # Convolutional Layer 1
    Conv2D(32, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),

    # Convolutional Layer 2
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),

    # Convolutional Layer 3
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),

    # Flatten the features
    Flatten(),

    # Fully Connected Dense Layer
    Dense(128, activation='relu'),
    Dropout(0.5),  # Dropout to prevent overfitting

    # Output Layer
    Dense(len(np.unique(labels)), activation='softmax')  # Softmax for multi-class classification
])

# Compile the model
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',  # For multi-class classification
    metrics=['accuracy']
)

# Summary of the model
model.summary()

# After training, you can evaluate the model on the test set
# test_loss, test_acc = model.evaluate(X_test, y_test)
# print(f"Test accuracy: {test_acc * 100:.2f}%")
```

```python
# Train the model
history = model.fit(
    X_train, y_train,
    validation_split=0.2,  # 20% of training data used for validation
    epochs=10,  # Adjust this based on your system and dataset size
    batch_size=32,
    verbose=1
)
```

```python
# Evaluate on the test dataset
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=1)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```

```python
# Save the model in Keras format
model.save('sign_language_model.keras')
print("Model saved in Keras format!")
```

```python
import cv2
import numpy as np
import mediapipe as mp
from tensorflow.keras.models import load_model
from sklearn.preprocessing import LabelEncoder
import pickle
import time

# Load the saved model
model = load_model('sign_language_model.keras')

# Load the saved label encoder
with open('label_encoder.pkl', 'rb') as f:
    label_encoder = pickle.load(f)

# Initialize MediaPipe Hand Tracking
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(min_detection_confidence=0.5, min_tracking_confidence=0.5)
mp_draw = mp.solutions.drawing_utils

# Start the webcam
cap = cv2.VideoCapture(0)

# Variables for controlling prediction frequency
last_pred_time = time.time()
pred_interval = 1  # seconds (predict every 1 second)
last_predicted_sign = ""  # Store the last predicted sign

# Start processing video frames
while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.flip(frame, 1)  # Flip horizontally
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = hands.process(rgb_frame)

    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            # Draw hand landmarks
            mp_draw.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)

            x_min = min([lm.x for lm in hand_landmarks.landmark])
            x_max = max([lm.x for lm in hand_landmarks.landmark])
            y_min = min([lm.y for lm in hand_landmarks.landmark])
            y_max = max([lm.y for lm in hand_landmarks.landmark])

            h, w, _ = frame.shape
            x_min, x_max = max(0, int(x_min * w)), min(w, int(x_max * w))
            y_min, y_max = max(0, int(y_min * h)), min(h, int(y_max * h))

            if x_max - x_min > 0 and y_max - y_min > 0:
                hand_roi = frame[y_min:y_max, x_min:x_max]
                resized_frame = cv2.resize(hand_roi, (64, 64))
                gray_frame = cv2.cvtColor(resized_frame, cv2.COLOR_BGR2GRAY)
                normalized_frame = gray_frame / 255.0
                reshaped_frame = normalized_frame.reshape(1, 64, 64, 1)

                # Check time for prediction
                if time.time() - last_pred_time > pred_interval:
                    prediction = model.predict(reshaped_frame)
                    predicted_label = np.argmax(prediction)
                    predicted_sign = label_encoder.inverse_transform([predicted_label])[0]

                    print(f"Predicted Sign: {predicted_sign}")

                    # Update the predicted sign and last prediction time
                    last_predicted_sign = predicted_sign
                    last_pred_time = time.time()

    # Display the predicted sign on the frame
    cv2.putText(frame, last_predicted_sign, (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)

    # Display the frame with the predicted sign text
    cv2.imshow('Sign Language Recognition', frame)

    # Exit if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
